```python
import requests
from bs4 import BeautifulSoup
import re
from string import *
```

```python

novel='  '
count=0

#定義a頁數
for page in range(1,17) :
    url = 'https://ielts-simon.com/ielts-help-and-english-pr/ielts-writing-task-2/page/'+ str(page)+'/'
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    articals = soup.find_all('div', re.compile('entry-body'))
    for tag in articals:  
        s = str(tag)
        words=s.split()
        if "band" in words:
            s_replace = s.replace('<br/>',"\n")
            while True:
                index_s = s_replace.find("<")
                index_e = s_replace.find(">",index_s+1)
                if index_s == -1:
                    break
                s_replace = s_replace.replace(s_replace[index_s:index_e+1],"")
            novel+=s_replace
            count+=1
            
print (novel) 
```

```python
with open("Simon-task2.txt", "w+") as my_file :

  my_file.write(novel)
```
